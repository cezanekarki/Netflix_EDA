```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drug_data = pd.read_csv("drug_data.csv")
drug_data
```

```python
fig = plt.figure(figsize=(25,6))
ax = fig.add_subplot()
ax.bar(drug_data["Country"],drug_data["Usage"])
ax.set_xlabel('Country')
ax.set_ylabel('Usage')
rects = ax.patches
labels = drug_data["Drug"]


for rect, label in zip(rects, labels):

    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, 10, label, ha="center", va="bottom", rotation = 90
    )


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

frequencies = drug_data["Usage"]

freq_series = pd.Series(frequencies)

x_labels = drug_data["Country"]

plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind="bar")
ax.set_xlabel("Countries")
ax.set_ylabel("Usage")
ax.set_xticklabels(x_labels)

rects = ax.patches

# Make some labels.
labels = drug_data["Drug"]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        (rect.get_x() + rect.get_width() / 2), 10, label, ha="center", va="bottom", rotation=90, color='black'
    )

plt.show()
```
