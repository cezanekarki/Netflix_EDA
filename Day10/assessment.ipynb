```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_data = pd.read_csv("full_grouped.csv")
```

```python
df_data.tail(10)
```

```python
lst = df_data["Country/Region"].unique()
```

```python
df_data["Country/Region"].nunique()
```

```python
country_choice = np.random.choice(lst)
```

```python
name_country = df_data[df_data["Country/Region"]==country_choice]
data_date = name_country["Date"]
#name_country.set_index = data_date
name_country.index = data_date
name_country
```

```python
ax = name_country.plot.line(figsize=(10,10),ylabel = "Number of Cases")
plt.title(country_choice)
```

```python
top_10_values = df_data.groupby(["Country/Region"]).sum().sort_values(by="Confirmed",ascending= False)
a =top_10_values.groupby(["Country/Region","Confirmed"]).first()
a.sort_values(by="Confirmed",ascending=False,inplace=True)
```

```python
a
```

```python
b= a.head(10)
```

```python
ax = b.plot(kind="bar",figsize=(10,10),xlabel="Countries",ylabel="Count")
```
