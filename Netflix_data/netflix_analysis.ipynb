```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix_data = pd.read_csv("netflix_titles.csv")
netflix_data.head()
```

```python
netflix_data.isna()
```

```python
ratings_uniques = netflix_data["rating"].unique()
ratings_uniques
```

```python
netflix_data["rating"].replace(to_replace=['min','nan'],regex=True,value=np.NaN, inplace=True)
ratings = netflix_data["rating"].unique()
netflix_data["rating"].fillna(netflix_data["rating"].value_counts().index[0],inplace=True)
netflix_data["rating"].isna()
```

```python
netflix_data.drop_duplicates(inplace=True)
netflix_data.dropna(inplace=True)
```

```python
#date added and release year gives different values. So formatting the date added to year
netflix_data["date_added"] = netflix_data["date_added"].str[-4:]
```

```python
#calculating the number of movies and tv shows
movies = netflix_data["type"].value_counts()
movies_df = pd.DataFrame(movies)
movies_df.transpose()
```

```python
top_directors = netflix_data["director"].value_counts().sort_values(ascending=False)
top_directors.head(10)
```

```python
#number of movies released in specific year

movies_year = netflix_data["release_year"].apply(pd.to_numeric).value_counts()
movies_year=movies_year.reset_index()
movies_year.columns=["Year","Count"]
movies_year.head(10)
```

```python
#counting the number of ratings
ratings_count = netflix_data["rating"].value_counts()
ratings_count = ratings_count.reset_index()
ratings_count.columns = ["Rating","Count"]
ratings_count
```

```python
netflix_data.head()
```

```python
#which kind of movie is most in netflix
listed_in = netflix_data[netflix_data["type"]=="Movie"]["listed_in"].str.split(", ")
genre_list = []
for i in listed_in:
    for j in i:
        genre_list.append(j)

genre = pd.DataFrame(genre_list)
genre.value_counts()
```

```python
#which kind of tv show is most in netflix
listed_tv = netflix_data[netflix_data["type"]=="TV Show"]["listed_in"].str.split(", ")
genre_tv = []
for i in listed_tv:
    for j in i:
        genre_tv.append(j)

genre = pd.DataFrame(genre_tv)
genre.value_counts()
```

```python
top_countries = netflix_data["country"].str.split(", ")
country_list = []
for i in top_countries:
    for j in i:
        country_list.append(j)

country = pd.DataFrame(country_list)
country.value_counts()
```

```python
casting_list = netflix_data["cast"].str.split(", ")
cast_new = []
for i in casting_list:
    for j in i:
        cast_new.append(j)

cast = pd.DataFrame(cast_new)
cast.value_counts()
```

```python
movies_duration = netflix_data[netflix_data["type"]=="Movie"]["duration"].value_counts()
movies_dr  = pd.DataFrame(movies_duration)
movies_dr.reset_index(inplace=True)
movies_dr.columns = ["Duration","Count"]
movies_dr
```

```python
tv_duration = netflix_data[netflix_data["type"]=="TV Show"]["duration"].value_counts()
tv_dr  = pd.DataFrame(tv_duration)
tv_dr.reset_index(inplace=True)
tv_dr.columns = ["Duration","Count"]
tv_dr
```
